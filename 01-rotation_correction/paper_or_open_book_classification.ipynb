```python
from os import walk

for (dirpath, dirnames, train_images) in walk('./data/train'):
    prefix_dir = lambda filename: './data/train/' + filename
    train_images = list(map(prefix_dir, train_images))

train_images
```

```python
from pigeon import annotate
from IPython.display import display, Image

annotations = annotate(
  train_images,
  options=['paper', 'open_book'],
  display_fn=lambda filename: display(Image(filename))
)
```
